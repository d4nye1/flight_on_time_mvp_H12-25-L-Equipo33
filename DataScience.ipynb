##Eliminación de columnas no útiles basadas en su utilidad para el MVP(producto mínimo viable)


*   Para el objetivo del proyecto (predecir retrasos), los vuelos cancelados o desviados no aportan valor, ya que no tienen tiempos reales confiables.




```python
df_clean = df[(df['cancelado'] == False) & (df['desviado'] == False)].copy()
```



*   Se eliminan vuelos que no completaron su operación normal, ya que el modelo debe aprender patrones de puntualidad en vuelos efectivamente realizados.



###Variables excluidas por fuga de información (data leakage)

```python
columnas_fuga = [
    'retraso_aerolinea',
    'retraso_clima',
    'retraso_trafico_aereo',
    'retraso_seguridad',
    'retraso_aeronave_previa',
    'codigo_cancelacion']

df_clean.drop(columns=columnas_fuga, inplace=True)
```



*   Las columnas no deben usarse, ya que explican directamente el retraso después de que ocurrió, contienen información posterior al evento y provocarían sobreajuste.



###Tratamiento de valores nulos


*   Observación esperada:
Los valores nulos restantes se concentran en:

Horarios reales

Retrasos

Tiempos operativos


```python
columnas_criticas = [
    'hora_salida_real',
    'hora_llegada_real',
    'retraso_salida',
    'retraso_llegada'
]

df_clean = df_clean.dropna(subset=columnas_criticas)
```



*   Para el análisis de retrasos, se requiere información completa de salida y llegada.



##Creación de variable objetivo


*   Se considera un vuelo retrasado si el retraso de salida es mayor a 15 minutos.

Esta información fue extraida de (https://www.oag.com/airline-on-time-performance-defining-late)



```python
df_clean['vuelo_retrasado'] = (df_clean['retraso_salida'] > 15).astype(int)
```

Valor	Significado
- 0	Puntual
- 1	Retrasado

##Componentes temporales

```python
df_clean['hora_salida'] = df_clean['hora_salida_programada'] // 100
df_clean['minuto_salida'] = df_clean['hora_salida_programada'] % 100
df_clean['fin_de_semana'] = df_clean['dia_semana'].isin([6, 7]).astype(int)
```

- Las variables temporales fueron transformadas en componentes más interpretables con el objetivo de capturar patrones operativos recurrentes. La hora de salida permite identificar acumulación de demoras a lo largo del día, mientras que la franja horaria agrupa horarios con comportamientos similares. Las variables de calendario como fin de semana y temporada del año permiten modelar ciclos semanales y estacionales sin introducir información posterior al evento. Estas transformaciones facilitan que el modelo aprenda patrones reales y aplicables en un entorno de predicción previa al despegue.

### Definición de la variable objetivo

El objetivo del proyecto es **predecir retrasos antes del despegue del vuelo**, utilizando únicamente información disponible **previa a la operación real**.

Por esta razón, la variable objetivo se define a partir del **retraso en la salida (`retraso_salida`)** y no del retraso en la llegada.  
El retraso en la llegada puede verse afectado por eventos ocurridos durante el vuelo (clima en ruta, congestión aérea posterior, decisiones operativas), los cuales no son observables ex-ante y podrían introducir ruido o fuga de información en el modelo.

La variable objetivo `vuelo_retrasado` se define como:

- **0 (Puntual):** retraso de salida ≤ 15 minutos  
- **1 (Retrasado):** retraso de salida > 15 minutos  

Este umbral se basa en el estándar de la industria para definir puntualidad operacional.

####Gráfica de distribución

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

retraso_counts = df_clean["vuelo_retrasado"].value_counts().sort_index()

labels = ['Puntual', 'Retrasado']
colors = ["#4CAF50", "#E74C3C"]  # Verde / Rojo


fig = make_subplots(
    rows=1,
    cols=2,
    specs=[[{"type": "bar"}, {"type": "pie"}]],
    subplot_titles=(
        "Distribución de Vuelos Puntuales y Retrasados",
        "Proporción de Vuelos Puntuales y Retrasados"
    )
)


# Gráfico de barras
fig.add_trace(
    go.Bar(
        x=labels,
        y=retraso_counts.values,
        text=[f"{v:,}" for v in retraso_counts.values],
        textposition="inside",
        marker_color=colors,
        textfont=dict(color="white", size=14),
        hovertemplate="<b>%{x}</b><br>Vuelos: %{y:,}<extra></extra>"
    ),
    row=1,
    col=1
)


# Gráfico de pastel

fig.add_trace(
    go.Pie(
        labels=labels,
        values=retraso_counts.values,
        hole=0,
        marker_colors=colors,
        textinfo="percent",
        textfont=dict(color="white", size=14),
        hovertemplate="<b>%{label}</b><br>%{percent}<br>Vuelos: %{value:,}<extra></extra>",
        sort=False
    ),
    row=1,
    col=2
)

fig.update_layout(
    template="plotly_white",
    showlegend=False,
    height=500,
    title_text="Análisis de la Variable Objetivo: Vuelo Retrasado",
    title_x=0.5
)

fig.show()

```

Interpretación de la Gráfica

- Se visualizó la distribución y proporción de la variable objetivo vuelo_retrasado mediante gráficos interactivos de barras y pastel. La representación permite identificar un desbalance moderado de clases, lo cual será considerado en la etapa de modelado predictivo.



###Retraso por aereolínias

```python
# Cálculo de proporciones
top_aerolineas = (
    df_clean.groupby('aerolinea')['vuelo_retrasado']
    .mean()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

fig = px.bar(
    top_aerolineas,
    x='aerolinea',
    y='vuelo_retrasado',
    text=top_aerolineas['vuelo_retrasado'].apply(lambda x: f"{x:.1%}"),
    color='vuelo_retrasado',
    color_continuous_scale=['#4CAF50', '#E74C3C'],
    title='Proporción de Vuelos Retrasados por Aerolínea (Top 10)'
)

fig.update_layout(
    xaxis_title='Aerolínea',
    yaxis_title='Probabilidad de retraso',
    yaxis_tickformat='.0%',
    coloraxis_showscale=False,
    template='simple_white'
)

fig.update_traces(
    textposition='outside'
)

fig.show()
```

Interpretación de la Gráfica

- Algunas aerolíneas presentan una probabilidad de retraso significativamente mayor que otras, lo cual indica patrones operativos que pueden ser aprendidos por el modelo predictivo.
