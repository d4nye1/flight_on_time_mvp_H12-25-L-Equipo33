```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import gdown
import zipfile
```

```python
url = 'https://drive.google.com/file/d/1Cr1qr3aDLuCrqlmldF7ywzxk4lyHUsgh/view?usp=sharing'
url = 'https://drive.google.com/uc?id=' + url.split('/')[-2]

output = 'flights_data.zip'
gdown.download(url, output, quiet=False)
```

```python
with zipfile.ZipFile("flights_data.zip", "r") as zip_ref:
    zip_ref.extractall("data")
```
